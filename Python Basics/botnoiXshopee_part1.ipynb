```
import pandas as pd
wordList = textList.split()
wordList
```

```
def getwordcount(wordList):
  wDic = {}
  for w in wordList:
    #print(w)
    if w in wDic.keys():
      wDic[w] = wDic[w]+1
    else:
      wDic[w] = 1
    #input()
  res = pd.DataFrame()
  res['vocab'] = wDic.keys()
  res['count'] = wDic.values()
  res = res.sort_values(by='count',ascending=False)
  return res
```

```
getwordcount(wordList)
```
